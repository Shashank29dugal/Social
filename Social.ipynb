```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df = df[["Age","EstimatedSalary","Purchased"]]
df
```

```python
df.info()
```

```python
purchased_data = df[["Age","EstimatedSalary","Purchased"]][df["Purchased"]==1]
purchased_data
```

```python
not_purchased_data = df[["Age","EstimatedSalary","Purchased"]][df["Purchased"]==0]
not_purchased_data
```

```python
plt.scatter(purchased_data["Age"],purchased_data["EstimatedSalary"],c="green")
plt.scatter(not_purchased_data["Age"],not_purchased_data["EstimatedSalary"],c="red")
```

```python
X = df[["Age","EstimatedSalary"]].values
y = df["Purchased"]
```

```python
X
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls = KNeighborsClassifier(n_neighbors= 7)
```

```python
cls.fit(X, y)
```

```python
cls.predict([[40,130000]])
```

```python
cls.predict([[20,130000]])
```

```python
cls.predict_proba([[40,130000]])
```
